```python
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib


plt.figure(figsize=(14, 10))
matplotlib.rcParams['font.family'] = ['Times New Roman']
matplotlib.rcParams['font.size'] = 45
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
label_size = 44
tick_size = 32
linewidth = 2.5
legend_size = 34
mark_size = 10
bwith = 2

x = np.array([1, 2, 3, 4, 5])
drl = [180, 180, 180, 180, 180]  
sxy = [50, 60, 101, 10, 47] 
yerr_drl = [2, 1, 2, 2, 1]   
yerr_sxy = [2, 2.3, 4.5, 1, 4] 


#plt.xlabel('Different Types of Workloads', fontsize=46)
plt.ylabel('Prediction overhead (ms)', fontsize=46)


plt.yticks(np.arange(0, 400, 50), fontsize=tick_size)

width = 0.3 
error_params = dict(elinewidth=3, ecolor='k', capsize=6, capthick=3)
plt.bar(x - width/2, drl, width, yerr=yerr_drl, error_kw=error_params, label='MBDP', color='w', hatch='x', edgecolor='#96b403', linewidth=3)
plt.bar(x + width/2, sxy, width, yerr=yerr_sxy, error_kw=error_params, label='U$^{2}$CMigration', color='w', hatch='xx', edgecolor='#db5856', linewidth=3)


plt.grid(axis='y')


x_labels = [u"ResNet50", u"YCSB", "Memcached", u"NPB-CG", u"Spark-Sort"]
plt.xticks(x, labels=x_labels,fontsize=tick_size)


plt.legend(loc=2, prop={'size': legend_size})

plt.savefig(os.path.join(".", "computeTime2.pdf"), format='pdf', bbox_inches='tight')
plt.savefig(os.path.join(".", "computeTime2.png"), format='png', bbox_inches='tight')
plt.show()
```
